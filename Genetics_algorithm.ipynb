```
import  numpy as np 
import random
import operator
```

```

#items = np.array([[15,15],[3,7],[2,10],[5,5],[9,8],[20,17]],dtype=float) #(wieght,value )
from google.colab import drive
import os 
drive.mount('/content/drive')
file=open('/content/drive/My Drive/CI_project/wieghts1.txt')
wieghts=[]
for i in  list(file):
    wieghts.append(int(i))
file=open('/content/drive/My Drive/CI_project/values1.txt')
values=[]
for i in  list(file):
    values.append(int(i))
    
```

```
items=np.array([wieghts,values])
items=items.T
print(items)
print(items.shape)
```

```
MAX_WIEGHT=6404180
```

```
class Individual :
    def __init__(self, gene,items,Max_wieght=MAX_WIEGHT):
            self.gene = gene 
            self.fitness=self.fit_func(items,Max_wieght)
    def __str__(self):
        return ("Individual: {} fitness: {}      wieght: {}".format(str(self.gene),self.fitness,self.wieght))
    def divide(self ,division_point=2 ): #divide array into two array in point array.size/devision_place
        return self.gene[:division_point],self.gene[division_point:]  
    def fit_func(self,items,Max_wieght):
        wieght=0
        value=0
        fitness=0 
        for i,g  in enumerate(self.gene):
            if g==1 :         
                wieght+= items[i][0]
                value+= items[i][1]
        self.wieght=wieght 
        if wieght > Max_wieght :
            fitness = 0
        else :
            fitness=value 
            
        return fitness
       
```

```
def Indvs_wieght(Indvs): # to comput probality of selecting each Indvs by its fitness
    s=1
    s=sum(i.fitness for i in Indvs)
    wieghts = list()
    for i in range(len(Indvs)) :
        wieghts.append(Indvs[i].fitness/s)
    return wieghts  

def select_parents(indvs,indvs_wieghts,number_of_parents=40): # Roulette Wheel Selection method  #number of selected  parent 
    return np.random.choice(indvs,size=number_of_parents,p=indvs_wieghts)






    
```

```
def generate_child(parents,items=items, Max_wieght=MAX_WIEGHT)    : #len( parents) must be even  
    division_point=np.random.randint(1,parents[0].gene.size-1)
    childs= list()
    for p in range(0, len(parents),2):
        mutasion=np.random.randn(1)*8
        x1,y1=parents[p].divide(division_point) # cross over 
        x2,y2=parents[p+1].divide(division_point)
        if mutasion <2:
            rangenetochange=np.random.randint(parents[0].gene.size-1)   #need better impelmention 
            if rangenetochange>x1.size:
                if y2[rangenetochange-x1.size]==0:
                    y2[rangenetochange-x1.size]=1
                else :
                    y2[rangenetochange-x1.size]=0
            else:
                if x1[rangenetochange-1]==0:
                    x1[rangenetochange-1]=1
                else :
                    x1[rangenetochange-1]=0
                
        childs.append(Individual(np.concatenate((x1,y2)),items,Max_wieght)) 
        childs.append(Individual(np.concatenate((x2,y1)),items,Max_wieght))
        
    return childs 

         
```

```
def make_init_pop(items,num_of_pop =80 ):
    num_of_items=items.shape[0]
    init_pop=list()
    for i in range(num_of_pop):
        init_pop.append(Individual(np.random.randint(2, size=(num_of_items,)),items=items))
    return init_pop 
    
```

```
def find_max(last_max,child):
    for i in child:
        if i.fitness>last_max.fitness:
            last_max=Individual(i.gene,items=items)
            print("-----------")
    return last_max
```

```
# selection is roulet wheel 
init_pop=make_init_pop(items,num_of_pop=2000)
parent_wieghts=Indvs_wieght(init_pop)
selected_parents=select_parents(init_pop , parent_wieghts,number_of_parents=1000)
child=generate_child(selected_parents)

parent_wieghts=Indvs_wieght(selected_parents)
child.extend(select_parents(selected_parents,parent_wieghts,number_of_parents=500))

num_of_items=items.shape[0]
Best=Individual(np.random.randint(2, size=(num_of_items,)),items=items)
Best=find_max(Best,child)
print(Best)
    

for i in range(2000):
    parent_wieghts=Indvs_wieght(child)
    selected_parents=select_parents(child , parent_wieghts,number_of_parents=1000)
    child=generate_child(selected_parents)

    parent_wieghts=Indvs_wieght(selected_parents)
    child.extend(select_parents(selected_parents,parent_wieghts,number_of_parents=500))
    Best=find_max(Best,child)
    print(Best)

    
```

```
def make_pointer (pop,n):# n=number of offspring to keep , pop should be sorted !
    F=sum(i.fitness for i in pop) #total fitness of population 
    P=F/n# distance between the pointers 
    start = np.random.random()
    pointers= [start+i*P for i in range(n)]
    return pointers
def select_parents_sus(pop,n ):
    Points=make_pointer (pop,n)
    keep=[]
    for P in Points:
        i=0
        while sum(pop[j].fitness for  j in range(i+1))<P:
            i+=1
        keep.append(pop[i])
    return keep
init_pop=make_init_pop(items,num_of_pop=100)
init_pop = sorted(init_pop, key=operator.attrgetter('fitness'))# note : after each generation you should sort the population by fitness

```

```
# selection is sus  
init_pop=make_init_pop(items,num_of_pop=1000)
init_pop = sorted(init_pop, key=operator.attrgetter('fitness'))

selected_parents=select_parents_sus(init_pop ,n=500)
child=generate_child(selected_parents)

parent_wieghts=Indvs_wieght(init_pop)
parent_wieghts=Indvs_wieght(selected_parents)
child.extend(select_parents(selected_parents,parent_wieghts,number_of_parents=100))

num_of_items=items.shape[0]
Best=Individual(np.random.randint(2, size=(num_of_items,)),items=items)
Best=find_max(Best,child)
print(Best)
    

for i in range(1000):
    child = sorted(child, key=operator.attrgetter('fitness'))
    parent_wieghts=Indvs_wieght(child)
    selected_parents=select_parents(child , parent_wieghts,number_of_parents=1000)
    child=generate_child(selected_parents)

    parent_wieghts=Indvs_wieght(selected_parents)
    child.extend(select_parents(selected_parents,parent_wieghts,number_of_parents=100))
    Best=find_max(Best,child)
    print(Best)

    
```
